```python
# Fungsi untuk menghitung efisiensi turbin
def hitung_efisiensi(input_energi, output_energi):
    # b. Mengembalikan nilai efisiensi
    efisiensi = output_energi / input_energi
    return efisiensi

# c. Contoh pemanggilan fungsi
hasil = hitung_efisiensi(500, 400)

# Menampilkan hasil efisiensi
print("Efisiensi turbin =", hasil * 100, "%")
```

Penjelasan

Program pada gambar tersebut digunakan untuk menghitung efisiensi turbin dengan menggunakan fungsi dalam Python. Fungsi bernama hitung_efisiensi memiliki dua parameter, yaitu input_energi sebagai energi masuk dan output_energi sebagai energi keluar dari turbin. Di dalam fungsi, efisiensi dihitung menggunakan rumus efisiensi = output_energi / input_energi, kemudian hasil perhitungan tersebut dikembalikan dengan perintah return. Setelah itu, fungsi dipanggil dengan nilai masukan input_energi = 500 dan output_energi = 400, sehingga diperoleh nilai efisiensi sebesar 0.8. Nilai ini kemudian dikalikan dengan 100 agar hasilnya dalam bentuk persen, dan ditampilkan menggunakan perintah print. Hasil akhirnya menunjukkan bahwa efisiensi turbin adalah 80,0%, yang berarti turbin mampu mengubah 80% energi masuk menjadi energi keluaran yang berguna.

```python
# Input data
tekanan_uap = float(input("Masukkan tekanan uap (psi): "))
status_manual_override = False  # contoh status manual override

# Logika if majemuk
if tekanan_uap > 150 and status_manual_override == False:
    print("Katup Terbuka")
else:
    print("Katup Tetap Tertutup")
```

Penjelasan

Program pada gambar tersebut berfungsi untuk menentukan kondisi katup (valve) berdasarkan tekanan uap dan status manual override. Pertama, program meminta pengguna memasukkan nilai tekanan uap dalam satuan psi melalui perintah input(), kemudian nilai tersebut diubah menjadi tipe data float agar bisa digunakan dalam perbandingan numerik. Variabel status_manual_override diatur bernilai False sebagai contoh, yang berarti sistem manual override tidak diaktifkan.

Selanjutnya, digunakan logika if majemuk dengan kondisi if tekanan_uap > 150 and status_manual_override == False:. Artinya, katup akan terbuka hanya jika tekanan uap lebih besar dari 150 psi dan manual override tidak aktif. Jika kedua syarat tersebut terpenuhi, program menampilkan pesan â€œKatup Terbukaâ€. Sebaliknya, jika salah satu kondisi tidak terpenuhi, maka program menampilkan â€œKatup Tetap Tertutupâ€.

```python
# Fungsi untuk menghitung efisiensi termal turbin gas
def hitung_efisiensi_termal(energi_masuk, energi_keluar):
    efisiensi = (energi_keluar / energi_masuk) * 100
    print("Efisiensi termal turbin gas =", efisiensi, "%")

# Contoh pemanggilan fungsi
hitung_efisiensi_termal(500, 400)
```

Penjelasan

Program pada gambar tersebut digunakan untuk menghitung efisiensi termal turbin gas dengan menggunakan fungsi dalam Python. Fungsi yang dibuat bernama hitung_efisiensi_termal, dan memiliki dua parameter yaitu energi_masuk serta energi_keluar
Setelah perhitungan dilakukan, hasil efisiensi ditampilkan menggunakan perintah print("Efisiensi termal turbin gas =", efisiensi, "%"). Pada bagian pemanggilan fungsi, digunakan nilai energi_masuk = 500 dan energi_keluar = 400
Sehingga program menampilkan output â€œEfisiensi termal turbin gas = 80.0 %â€, yang berarti turbin gas bekerja dengan efisiensi sebesar 80 persen dari energi panas yang diterimanya.

```python
# Inisialisasi volume awal tangki
tangki = 0

# Loop pengisian bahan bakar
while tangki < 100:
    tangki += 10  # Naikkan 10 liter setiap iterasi
    print("Volume saat ini:", tangki, "liter")

    # Peringatan saat hampir penuh
    if tangki == 80:
        print("WARNING: Hampir penuh!")

    # Pesan saat tangki penuh
    if tangki == 100:
        print("TANGKI PENUH â€“ Pengisian berhenti.")
```

Penjelasan

Program pada gambar tersebut digunakan untuk mensimulasikan proses pengisian tangki bahan bakar hingga penuh dengan menambahkan volume secara bertahap.

Pertama, variabel tangki diinisialisasi dengan nilai 0, yang berarti tangki dalam keadaan kosong. Kemudian digunakan perulangan while tangki < 100, artinya program akan terus berjalan selama volume tangki masih kurang dari 100 liter. Di dalam loop, setiap iterasi menambah volume tangki sebesar 10 liter menggunakan tangki += 10, lalu menampilkan volume saat ini dengan perintah print("Volume saat ini:", tangki, "liter").
